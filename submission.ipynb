```python
import pandas as pd
face =  pd.read_csv("_face_new/face_solution.csv", sep=';', index_col='ImageID')
head =  pd.read_csv("_head_new/head_solution.csv", sep=';', index_col='ImageID')
carplate =  pd.read_csv("_carplate/carplate_solution.csv", sep=';', index_col='ImageID')
human =  pd.read_csv("_human/human_solution.csv", sep=';', index_col='ImageID')
#переделать
car =  pd.read_csv("_car/car_solution.csv", sep=';', index_col='ImageID')
```

```python
def rescale(df):
    #x_c, y_c, w, h
    df['x_c']=df['XMin']
    df['y_c']=df['XMax']
    df['w']=df['YMin']
    df['h']=df['YMax']
    df['XMin'] = df['x_c'] - 0.5 * df['w']
    df['XMax'] = df['x_c'] + 0.5 * df['w']
    df['YMin'] = df['y_c'] - 0.5 * df['h']
    df['YMax'] = df['y_c'] + 0.5 * df['h']
    return df[['LabelName','Conf','XMin','XMax','YMin','YMax']]
```

```python

carplate = rescale(carplate)
carplate.head()
##xmin, ymin, xmax, ymax
##xmin, xmax, ymin, ymax
```

```python
human = rescale(human)
human.head()
```

```python
car = rescale(car)
car.head()
```

```python
sub = pd.concat([car, head,face,human,carplate], axis=0)
```

```python
sub_filter = sub[sub['Conf'] > 0.4]
```

```python
sub_filter.to_csv('submission.csv',sep=';')
```
